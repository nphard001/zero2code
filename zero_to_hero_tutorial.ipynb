## 第三章：CLI 是什麼鬼？

### 📱 類比說明

想像你的電腦有兩種操作方式：

#### 1. GUI（圖形介面）- 你平常用的
- 用滑鼠點擊
- 看得到圖示和按鈕
- 像是用遙控器看電視

#### 2. CLI（命令列介面）- 現在在學的
- 用鍵盤打字
- 只有文字
- 像是傳簡訊給電腦

### 🎯 為什麼要學 CLI？

因為 Claude Code 就是透過 CLI 來工作的！

**舉例：**
- GUI：點擊「新增資料夾」按鈕
- CLI：輸入 `mkdir 資料夾名稱`

結果一樣，但 CLI 可以：
- 更快（熟練後）
- 自動化
- 讓 Claude Code 幫你做事

## 第四章：Git 到底是什麼？

### 🎮 用遊戲來理解 Git

還記得我在第一個教學說 Git 像遊戲存檔嗎？讓我們更深入：

**沒有 Git 的世界：**
```
報告_v1.doc
報告_v2.doc
報告_v2_final.doc
報告_v2_final_真的final.doc
報告_v2_final_真的final_改過.doc
```

**有 Git 的世界：**
```
報告.doc
（Git secretly 記錄所有版本）
```

### 🔄 Git 的工作流程

當 Claude Code 幫你執行 Git 命令時，實際上是在做：

1. **git init** = 「開始記錄這個資料夾的變化」
2. **git add** = 「把這些改變準備存檔」
3. **git commit** = 「正式存檔並寫下這次改了什麼」
4. **git log** = 「看看存檔記錄」

## 第八章：常見困惑解答

### ❓ 我打的命令去哪了？

在 CLI 中：
- **向上箭頭** ↑ - 顯示上一個命令
- **向下箭頭** ↓ - 顯示下一個命令
- 所以你可以重複使用之前的命令

### ❓ 為什麼有時候沒反應？

可能原因：
1. Claude Code 正在思考（等一下）
2. 命令正在執行（看看有沒有輸出）
3. 等待你的輸入（檢查是否有提示）

### ❓ 我怎麼知道命令成功了？

- **沒消息就是好消息** - 很多命令成功時不會說話
- **看提示符號** - 回到 `$` 或 `>` 表示完成
- **用 git status** - 隨時檢查狀態

### ❓ 我可以刪除或修改嗎？

當然可以！
- 刪除檔案：`rm 檔名`
- 刪除資料夾：`rm -r 資料夾名`
- 但 Git 會記錄一切，所以不用怕

### 🎯 Git 必備命令

```bash
日常使用：
git status          # 看狀態
git add .           # 加入所有改變
git commit -m "訊息" # 提交
git log --oneline   # 看歷史

進階一點：
git diff            # 看改了什麼
git branch          # 看分支
git checkout -b 新分支 # 建立新分支
```

### 🎯 CLI 生存指令

```bash
導航：
pwd                 # 我在哪裡？
ls                  # 這裡有什麼？
cd 資料夾            # 進入資料夾
cd ..               # 回上一層

檔案操作：
touch 檔名.txt       # 建立檔案
mkdir 資料夾名       # 建立資料夾
rm 檔名             # 刪除檔案
cp 來源 目標         # 複製
mv 來源 目標         # 移動/重新命名
```
